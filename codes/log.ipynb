```python
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt

def log_transformation(image_array, c=1):
    
    # Apply log transformation   Formula: s = c * log(1 + r)
    log_image = c * np.log(1 + image_array)

    # Normalize to range [0, 255]. To normalize divide each pixel of image by maximum value of image * and scale to 255 range.  
    # so valus in 0-1 range will come to range 0-255
    log_image = (log_image / log_image.max()) * 255
    return log_image.astype(np.uint8)


# Main function   # Load the image and convert to grayscale
if __name__ == "__main__":
    # # Replace with your image path
    image_path = "C:/Users/gaura/Downloads/logTranformImage.jpg"

    img = Image.open(image_path).convert('L')  # Convert to grayscale
    img_array = np.array(img)

    # Perform log transformation
    log_image = log_transformation(img_array, c=1)

   
    # Display the results
    plt.figure(figsize=(18, 6))

    plt.subplot(1, 3, 1)
    plt.imshow(img_array, cmap='gray')
    plt.title('Original Image')
    plt.axis('off')

    plt.subplot(1, 3, 2)
    plt.imshow(log_image, cmap='gray')
    plt.title('Log Transformation')
    plt.axis('off')

    plt.tight_layout()
    plt.show()
```
