```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image

# Load the grayscale image
image_path = "C:/Users/gaura/Downloads/pngimg.com - dot_PNG40.png"
image = Image.open(image_path).convert("L")
image_np = np.array(image)

# Negative Image
negative_image = 255 - image_np

# Display the original and negative images
plt.figure(figsize=(8, 4))

plt.subplot(1, 2, 1)
plt.imshow(image_np, cmap='gray')
plt.title("Original Image")
plt.axis("off")

plt.subplot(1, 2, 2)
plt.imshow(negative_image, cmap='gray')
plt.title("Negative Image")
plt.axis("off")

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image

# Load the grayscale image
image_path = "C:/Users/gaura/Downloads/dog.jfif"
image = Image.open(image_path).convert("L")
image_np = np.array(image)

# Thresholding
threshold_value = 128
threshold_image = np.where(image_np > threshold_value, 255, 0).astype(np.uint8)

# Display the original and thresholded images
plt.figure(figsize=(8, 4))

plt.subplot(1, 2, 1)
plt.imshow(image_np, cmap='gray')
plt.title("Original Image")
plt.axis("off")

plt.subplot(1, 2, 2)
plt.imshow(threshold_image, cmap='gray')
plt.title("Thresholded Image")
plt.axis("off")

plt.tight_layout()
plt.show()
```
